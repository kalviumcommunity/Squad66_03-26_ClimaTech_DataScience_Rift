```python
import pandas as pd
import numpy as np

print("----- 1. Understanding Pandas Series -----")

# Create a simple Series
s = pd.Series([10, 20, 30, 40])
print("Series:\n", s)

print("Values:", s.values)
print("Index:", s.index)


print("\n----- 2. Creating Series from Python List -----")

data_list = [5, 15, 25, 35]

series_from_list = pd.Series(data_list)
print("Series from List:\n", series_from_list)

print("Values:", series_from_list.values)
print("Index:", series_from_list.index)


print("\n----- 3. Creating Series with Custom Index -----")

series_custom_index = pd.Series(data_list, index=['a', 'b', 'c', 'd'])
print("Series with Custom Index:\n", series_custom_index)

# Correct access methods
print("Access by label (b):", series_custom_index['b'])
print("Access by position (0):", series_custom_index.iloc[0])


print("\n----- 4. Creating Series from NumPy Array -----")

np_array = np.array([100, 200, 300])

series_from_array = pd.Series(np_array)
print("Series from NumPy Array:\n", series_from_array)

print("Values:", series_from_array.values)
print("Index:", series_from_array.index)


print("\n----- 5. Basic Operations on Series -----")

s2 = pd.Series([1, 2, 3, 4])

print("Original Series:\n", s2)
print("Add 10:\n", s2 + 10)
print("Multiply by 2:\n", s2 * 2)


print("\n----- 6. Why Series is Useful -----")

print("Series has labels (index) + values")
print("This makes data more meaningful compared to NumPy arrays")

print("\n----- Pandas Series Demo Completed -----")
```
