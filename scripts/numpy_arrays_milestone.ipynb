```python
import numpy as np
```

```python
list1=[1,2,3,4,5]
arr1=np.array(list1)
print(arr1)
```

```python
list2 = [[1, 2, 3],
         [4, 5, 6]]

arr2 = np.array(list2)

print(arr2)
```

```python
print(arr1.shape)   # shape (dimensions)
print(arr1.dtype)   # data type

print(arr2.shape)
print(arr2.dtype)
```

```python
arr = np.array([10, 20, 30])

print(arr + 5)   # add 5 to each element
print(arr * 2) 
```
