## 7. Histogram Strategy Guide

```python
print("MISTAKE 1: Too few bins loses detail")
data = np.random.normal(100, 15, 1000)
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].hist(data, bins=3, edgecolor='black', alpha=0.7, color='red')
axes[0].set_title('Too Few Bins (3) - Too Coarse', fontweight='bold')
axes[0].set_ylabel('Frequency')

axes[1].hist(data, bins=20, edgecolor='black', alpha=0.7, color='green')
axes[1].set_title('Good Bins (20) - Balanced', fontweight='bold')
axes[1].set_ylabel('Frequency')

axes[2].hist(data, bins=100, edgecolor='black', alpha=0.7, color='blue')
axes[2].set_title('Too Many Bins (100) - Too Noisy', fontweight='bold')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

print("→ Too few bins: Can't see pattern")
print("→ Too many bins: Noise, hard to interpret")
print("→ Good rule: 20-30 bins for most data\n" + "="*60 + "\n")

print("MISTAKE 2: Using histogram for categorical data")
categories = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Red', 'Green', 'Blue', 'Red', 'Green', 'Blue']
})
print("Data:", categories['Color'].tolist())
print("❌ This is categorical - DON'T use histogram!")
print("✓ Use bar chart or value_counts() instead:")
print(categories['Color'].value_counts())
print("\n" + "="*60 + "\n")

print("MISTAKE 3: Comparing histograms with different scales")
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

small_values = np.random.normal(10, 2, 100)
large_values = np.random.normal(5000, 1000, 100)

axes[0].hist(small_values, bins=20, edgecolor='black', alpha=0.7, color='red')
axes[0].set_title('Small Values (0-20)')
axes[0].set_ylabel('Frequency')

axes[1].hist(large_values, bins=20, edgecolor='black', alpha=0.7, color='blue')
axes[1].set_title('Large Values (3000-7000)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

print("❌ Can't compare directly - different scales!")
print("✓ Normalize by dividing by standard deviation first\n" + "="*60 + "\n")

print("MISTAKE 4: Misinterpreting small datasets")
tiny = np.array([1, 2, 5, 8, 9])
plt.figure(figsize=(10, 5))
plt.hist(tiny, bins=5, edgecolor='black', alpha=0.7, color='orange')
plt.title('Histogram of Only 5 Values - Not Meaningful')
plt.ylabel('Frequency')
plt.show()
print("❌ Only 5 data points - histogram is unreliable")
print("✓ Need at least 30-50 values for meaningful histograms")
```

## 6. Common Mistakes to Avoid

```python
store_sales = pd.DataFrame({
    'Store A': np.random.normal(1000, 50, 100),
    'Store B': np.random.normal(1000, 200, 100),
    'Store C': np.random.normal(900, 100, 100)
})

print("Store Sales Data:")
print(f"Mean: A={store_sales['Store A'].mean():.0f}, B={store_sales['Store B'].mean():.0f}, C={store_sales['Store C'].mean():.0f}")
print(f"Std:  A={store_sales['Store A'].std():.0f}, B={store_sales['Store B'].std():.0f}, C={store_sales['Store C'].std():.0f}")

print("\n" + "="*60)
print("\nMethod 1: Subplots (one histogram per column)")
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for idx, col in enumerate(store_sales.columns):
    axes[idx].hist(store_sales[col], bins=15, edgecolor='black', alpha=0.7, color=['blue', 'orange', 'green'][idx])
    axes[idx].set_title(f'{col}', fontweight='bold')
    axes[idx].set_xlabel('Sales ($)')
    axes[idx].set_ylabel('Frequency')
    axes[idx].axvline(store_sales[col].mean(), color='red', linestyle='--', linewidth=2)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("\nMethod 2: Side-by-side on same plot (overlapping)")
fig, ax = plt.subplots(figsize=(12, 6))

for col in store_sales.columns:
    ax.hist(store_sales[col], bins=15, alpha=0.5, label=col, edgecolor='black')

ax.set_title('Store Sales Comparison', fontweight='bold', fontsize=12)
ax.set_xlabel('Sales ($)', fontsize=10)
ax.set_ylabel('Frequency', fontsize=10)
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.show()

print("\nComparison insights:")
print("✓ Store A: Narrow range (consistent sales)")
print("✓ Store B: Wide range (unpredictable sales)")
print("✓ Store C: Slightly lower average than A &amp; B")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

print("Different Distribution Shapes:\n")

print("Shape 1: NORMAL (Bell Curve)")
print("  Most values cluster around center")
print("  Symmetric, tapers evenly on both sides")
normal_data = np.random.normal(100, 15, 1000)
axes[0, 0].hist(normal_data, bins=30, edgecolor='black', alpha=0.7, color='green')
axes[0, 0].set_title('Normal Distribution', fontweight='bold')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].axvline(np.mean(normal_data), color='red', linestyle='--', linewidth=2)
print(f"  Mean ≈ Median ≈ {np.mean(normal_data):.0f}\n")

print("Shape 2: RIGHT-SKEWED (Tail on right)")
print("  Long tail extending to the right")
print("  Most values on left, few high outliers")
right_skewed = np.random.exponential(30, 1000)
axes[0, 1].hist(right_skewed, bins=30, edgecolor='black', alpha=0.7, color='orange')
axes[0, 1].set_title('Right-Skewed Distribution', fontweight='bold')
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].axvline(np.mean(right_skewed), color='red', linestyle='--', linewidth=2, label='Mean')
axes[0, 1].axvline(np.median(right_skewed), color='blue', linestyle='--', linewidth=2, label='Median')
axes[0, 1].legend()
print(f"  Mean > Median (mean pulled right by outliers)\n")

print("Shape 3: LEFT-SKEWED (Tail on left)")
print("  Long tail extending to the left")
print("  Most values on right, few low outliers")
left_skewed = 100 - np.random.exponential(30, 1000)
axes[1, 0].hist(left_skewed, bins=30, edgecolor='black', alpha=0.7, color='purple')
axes[1, 0].set_title('Left-Skewed Distribution', fontweight='bold')
axes[1, 0].set_xlabel('Value')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].axvline(np.mean(left_skewed), color='red', linestyle='--', linewidth=2, label='Mean')
axes[1, 0].axvline(np.median(left_skewed), color='blue', linestyle='--', linewidth=2, label='Median')
axes[1, 0].legend()
print(f"  Mean < Median (mean pulled left by outliers)\n")

print("Shape 4: BIMODAL (Two peaks)")
print("  Two distinct clusters/groups")
print("  Two separate peak areas")
bimodal = np.concatenate([np.random.normal(40, 5, 500), np.random.normal(80, 5, 500)])
axes[1, 1].hist(bimodal, bins=30, edgecolor='black', alpha=0.7, color='teal')
axes[1, 1].set_title('Bimodal Distribution', fontweight='bold')
axes[1, 1].set_xlabel('Value')
axes[1, 1].set_ylabel('Frequency')
print(f"  Two separate groups mixing in data\n")

plt.tight_layout()
plt.show()
```
