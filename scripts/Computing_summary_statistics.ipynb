## 10. When to Use Which Statistic

```python
print("MISTAKE 1: Using mean without checking for outliers")
data_with_outlier = pd.DataFrame({'Value': [10, 12, 11, 9, 100]})
print(f"Data: {data_with_outlier['Value'].tolist()}")
print(f"Mean: {data_with_outlier['Value'].mean():.1f} ← Skewed by 100!")
print(f"Median: {data_with_outlier['Value'].median():.1f} ← More representative")
print("→ Always check median too\n" + "="*60 + "\n")

print("MISTAKE 2: Ignoring std deviation")
group1 = pd.DataFrame({'Score': [50, 50, 50, 50, 50]})
group2 = pd.DataFrame({'Score': [0, 25, 50, 75, 100]})
print(f"Group 1 scores: {group1['Score'].tolist()}")
print(f"Group 1 mean: {group1['Score'].mean():.1f}, std: {group1['Score'].std():.1f}")
print(f"\nGroup 2 scores: {group2['Score'].tolist()}")
print(f"Group 2 mean: {group2['Score'].mean():.1f}, std: {group2['Score'].std():.1f}")
print("→ Same mean, very different spread!")
print("→ Always look at std dev\n" + "="*60 + "\n")

print("MISTAKE 3: Not considering context")
test_data = pd.DataFrame({'Score': [0, 50, 95]})
print(f"Scores: {test_data['Score'].tolist()}")
print(f"Mean: {test_data['Score'].mean():.1f}")
print("❌ Not enough data to conclude anything (only 3 values)")
print("❌ Need to know: What's the max score? What's passing?")
print("→ Always understand what the numbers represent")
```

## 7. Comparing Statistics Across Columns

```python
website_traffic = pd.DataFrame({
    'Daily Visitors': [1000, 1100, 950, 1050, 15000, 1020, 980]
})

print("Website Daily Visitors:")
print(website_traffic)
print("\n" + "="*60)

mean = website_traffic['Daily Visitors'].mean()
median = website_traffic['Daily Visitors'].median()
std = website_traffic['Daily Visitors'].std()

print(f"\nMean: {mean:,.0f}")
print(f"Median: {median:,.0f}")
print(f"Std Dev: {std:,.0f}")
print(f"Min: {website_traffic['Daily Visitors'].min():,.0f}")
print(f"Max: {website_traffic['Daily Visitors'].max():,.0f}")

print("\n" + "="*60)
print("\nOutlier Detection using IQR (Interquartile Range):")
Q1 = website_traffic['Daily Visitors'].quantile(0.25)
Q3 = website_traffic['Daily Visitors'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Q1 (25%): {Q1:,.0f}")
print(f"Q3 (75%): {Q3:,.0f}")
print(f"IQR: {IQR:,.0f}")
print(f"Outlier bounds: {lower_bound:,.0f} to {upper_bound:,.0f}")

outliers = website_traffic[(website_traffic['Daily Visitors'] < lower_bound) | 
                           (website_traffic['Daily Visitors'] > upper_bound)]
print(f"\nOutliers found: {outliers['Daily Visitors'].tolist()}")
print("→ 15000 visitors is unusually high!")
```
