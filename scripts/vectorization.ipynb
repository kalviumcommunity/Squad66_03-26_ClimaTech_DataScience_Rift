```python
import numpy as np

print("----- 1. Loop vs Vectorized Example -----")

# Create array
arr = np.array([1, 2, 3, 4, 5])

# Loop-based approach
loop_result = []
for i in range(len(arr)):
    loop_result.append(arr[i] * 2)

print("Loop Result:", loop_result)

# Vectorized approach
vector_result = arr * 2
print("Vectorized Result:", vector_result)


print("\n----- 2. Vectorized Arithmetic Operations -----")

arr2 = np.array([10, 20, 30])

print("Original Array:", arr2)
print("Add 5:", arr2 + 5)
print("Multiply by 2:", arr2 * 2)
print("Divide by 2:", arr2 / 2)


print("\n----- 3. Vectorized Comparisons -----")

arr3 = np.array([10, 20, 30, 40])

comparison = arr3 > 25
print("Array:", arr3)
print("Elements greater than 25:", comparison)


print("\n----- 4. Filtering using Conditions -----")

filtered = arr3[arr3 > 20]
print("Filtered elements (greater than 20):", filtered)


print("\n----- 5. Broadcasting Example -----")

arr4 = np.array([1, 2, 3])

# Broadcasting: adding scalar to array
print("Array + 10:", arr4 + 10)


print("\n----- 6. Common Mistake (Shape Mismatch) -----")

a = np.array([1, 2, 3])
b = np.array([1, 2])

try:
    print(a + b)
except ValueError as e:
    print("Error due to shape mismatch:", e)


print("\n----- Completed Vectorization Demo -----")
```
