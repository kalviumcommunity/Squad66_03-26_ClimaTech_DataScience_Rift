```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'StudentID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', np.nan, 'Charlie', 'Diana', np.nan, 'Frank', 'Grace', 'Henry', 'Iris', 'Jack'],
    'Age': [20, 21, np.nan, 22, 23, np.nan, 21, 20, 22, 21],
    'Score': [85, 90, 88, np.nan, 92, 78, np.nan, 84, 91, 87],
    'Grade': ['B', 'A', 'B', 'B', 'A', np.nan, 'A', 'B', 'A', 'B'],
    'Attendance': [0.95, np.nan, 0.88, 0.92, np.nan, 0.85, 0.98, 0.90, np.nan, 0.93]
})

print("Original DataFrame:")
print(df)
print("\n" + "="*60)
print(f"Shape: {df.shape}")
print(f"Data types:\n{df.dtypes}")
print("\n" + "="*60)
print(f"Missing values per column:\n{df.isnull().sum()}")
```

```python
print("Drop rows with missing values in specific columns:")
df_drop_specific = df.dropna(subset=['Score', 'Age'])
print(df_drop_specific)
print(f"Original shape: {df.shape}")
print(f"After dropping rows with missing Score or Age: {df_drop_specific.shape}")
print("\n" + "="*60 + "\n")

print("Drop columns with missing values:")
df_drop_cols = df.dropna(axis=1)
print(df_drop_cols)
print(f"Original columns: {df.shape[1]}")
print(f"Remaining columns: {df_drop_cols.shape[1]}")
print(f"Columns removed: {df.shape[1] - df_drop_cols.shape[1]}")
```

```python
print("Fill missing values with column median:")
df_fill_median = df.copy()
df_fill_median['Score'] = df_fill_median['Score'].fillna(df_fill_median['Score'].median())
df_fill_median['Age'] = df_fill_median['Age'].fillna(df_fill_median['Age'].median())
print(df_fill_median)
print("\n" + "="*60 + "\n")

print("Fill missing values with column mode (most frequent):")
df_fill_mode = df.copy()
df_fill_mode['Grade'] = df_fill_mode['Grade'].fillna(df_fill_mode['Grade'].mode()[0])
print(df_fill_mode)
```

```python
print("Strategy 3: SELECTIVE - Drop problematic columns, fill others")
df_strategy_selective = df.copy()
df_strategy_selective = df_strategy_selective.drop('Attendance', axis=1)
df_strategy_selective['Score'] = df_strategy_selective['Score'].fillna(df_strategy_selective['Score'].mean())
df_strategy_selective['Age'] = df_strategy_selective['Age'].fillna(df_strategy_selective['Age'].mean())
df_strategy_selective['Grade'] = df_strategy_selective['Grade'].fillna(df_strategy_selective['Grade'].mode()[0])
df_strategy_selective['Name'] = df_strategy_selective['Name'].fillna('Unknown')
print(f"Shape after selective handling: {df_strategy_selective.shape}")
print("Data retained:")
print(df_strategy_selective)
```

```python
print("Exercise 4: Drop column with highest missing percentage")
missing_pct = (df.isnull().sum() / len(df)) * 100
print("Missing percentages:")
print(missing_pct)
drop_col = missing_pct.idxmax()
result_4 = df.drop(drop_col, axis=1)
print(f"\nDropped column: {drop_col}")
print(result_4)
print("\n" + "="*60 + "\n")

print("Exercise 5: Fill numeric columns with mean, categorical with mode")
result_5 = df.copy()
result_5['Score'] = result_5['Score'].fillna(result_5['Score'].mean())
result_5['Age'] = result_5['Age'].fillna(result_5['Age'].mean())
result_5['Attendance'] = result_5['Attendance'].fillna(result_5['Attendance'].mean())
result_5['Grade'] = result_5['Grade'].fillna(result_5['Grade'].mode()[0])
result_5['Name'] = result_5['Name'].fillna('Unknown')
print(result_5)
print(f"Total missing after fill: {result_5.isnull().sum().sum()}")
```

```python
print("Exercise 1: Drop rows where Score is missing")
result_1 = df.dropna(subset=['Score'])
print(result_1)
print(f"Rows remaining: {result_1.shape[0]}")
print("\n" + "="*60 + "\n")

print("Exercise 2: Fill Age with median")
result_2 = df.copy()
result_2['Age'] = result_2['Age'].fillna(result_2['Age'].median())
print(result_2)
print(f"Missing Age after fill: {result_2['Age'].isnull().sum()}")
print("\n" + "="*60 + "\n")

print("Exercise 3: Fill Name with 'Unknown' and Grade with mode")
result_3 = df.copy()
result_3['Name'] = result_3['Name'].fillna('Unknown')
result_3['Grade'] = result_3['Grade'].fillna(result_3['Grade'].mode()[0])
print(result_3)
print(f"Missing Name: {result_3['Name'].isnull().sum()}")
print(f"Missing Grade: {result_3['Grade'].isnull().sum()}")
```

## 8. Decision Guide: When to Drop vs Fill

```python
print("MISTAKE 1: Dropping without knowing the impact")
print(f"Original rows: {df.shape[0]}")
print(f"After dropna(): {df.dropna().shape[0]}")
print(f"Data loss: {df.shape[0] - df.dropna().shape[0]} rows ({((df.shape[0] - df.dropna().shape[0])/df.shape[0]*100):.1f}%)")
print("Result: Lost 6 rows (60% of data!) - Probably too aggressive for this dataset")
print("\n" + "="*60 + "\n")

print("MISTAKE 2: Filling numeric data into categorical columns")
df_mistake = df.copy()
df_mistake['Grade'] = df_mistake['Grade'].fillna(0)
print("After filling Grade with numeric 0:")
print(df_mistake[['StudentID', 'Grade']])
print("Problem: Grade should be 'A', 'B', etc., not 0")
print("\n" + "="*60 + "\n")

print("MISTAKE 3: Filling without considering the meaning")
df_mistake = df.copy()
df_mistake['Score'] = df_mistake['Score'].fillna(0)
print("After filling Score with 0:")
print(df_mistake[['StudentID', 'Score']])
print("Problem: Score of 0 implies student failed, not missing data")
```

## 7. Common Mistakes to Avoid

```python
print("Strategy 1: DROP all rows with any missing")
df_strategy_drop = df.dropna()
print(f"Shape after dropping: {df_strategy_drop.shape}")
print("Data retained:")
print(df_strategy_drop)
print("\n" + "="*60 + "\n")

print("Strategy 2: FILL with mean/median/mode")
df_strategy_fill = df.copy()
df_strategy_fill['Score'] = df_strategy_fill['Score'].fillna(df_strategy_fill['Score'].mean())
df_strategy_fill['Age'] = df_strategy_fill['Age'].fillna(df_strategy_fill['Age'].mean())
df_strategy_fill['Grade'] = df_strategy_fill['Grade'].fillna(df_strategy_fill['Grade'].mode()[0])
df_strategy_fill['Name'] = df_strategy_fill['Name'].fillna('Unknown')
df_strategy_fill['Attendance'] = df_strategy_fill['Attendance'].fillna(df_strategy_fill['Attendance'].mean())
print(f"Shape after filling: {df_strategy_fill.shape}")
print("Data retained:")
print(df_strategy_fill)
```

## 6. Comparing Drop vs Fill Strategies

```python
df_temporal = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=8),
    'Temperature': [20, np.nan, 22, np.nan, 25, 24, np.nan, 26],
    'Humidity': [60, 65, np.nan, 70, np.nan, 72, 71, np.nan]
})

print("Original temporal data:")
print(df_temporal)
print("\n" + "="*60 + "\n")

print("Forward fill (use previous value):")
df_ffill = df_temporal.fillna(method='ffill')
print(df_ffill)
print("\n" + "="*60 + "\n")

print("Backward fill (use next value):")
df_bfill = df_temporal.fillna(method='bfill')
print(df_bfill)
```
