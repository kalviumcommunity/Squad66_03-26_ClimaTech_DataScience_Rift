```python
import pandas as pd

print("----- 1. Understanding Pandas DataFrame -----")

# Create a simple DataFrame
data = {
    "Name": ["John", "Alice", "Bob"],
    "Age": [25, 30, 22],
    "City": ["Delhi", "Mumbai", "Chandigarh"]
}

df = pd.DataFrame(data)

print("DataFrame:\n", df)
print("Shape (rows, columns):", df.shape)
print("Columns:", df.columns)
print("Data Types:\n", df.dtypes)


print("\n----- 2. Creating DataFrame from Dictionary -----")

students = {
    "Name": ["Aman", "Simran", "Rohit"],
    "Marks": [85, 90, 78],
    "Grade": ["A", "A+", "B"]
}

df_students = pd.DataFrame(students)

print("Students DataFrame:\n", df_students)


print("\n----- 3. Creating DataFrame from CSV File -----")

# First, create a sample CSV file
sample_data = {
    "Product": ["Laptop", "Phone", "Tablet"],
    "Price": [70000, 30000, 20000],
    "Stock": [10, 25, 15]
}

sample_df = pd.DataFrame(sample_data)

# Save to CSV
sample_df.to_csv("sample_data.csv", index=False)

# Load from CSV
df_file = pd.read_csv("sample_data.csv")

print("Loaded DataFrame from CSV:\n", df_file)


print("\n----- 4. Inspecting DataFrame -----")

print("First 5 rows:\n", df_file.head())
print("Last 5 rows:\n", df_file.tail())

print("Shape:", df_file.shape)
print("Columns:", df_file.columns)
print("Data Types:\n", df_file.dtypes)


print("\n----- 5. Why DataFrames are Useful -----")

print("DataFrames store data in rows and columns (like Excel)")
print("They make data easy to read, analyze, and manipulate")


print("\n----- DataFrame Demo Completed -----")
```
