```python
student_grades = pd.DataFrame({
    'Math': [75, 82, 78, 85, 88, 80, 92, 79, 86, 81, 77],
    'Science': [70, 88, 75, 90, 92, 85, 95, 78, 89, 84, 72],
    'History': [80, 79, 82, 81, 80, 83, 81, 80, 82, 79, 81]
})

print("Student Grades by Subject:")
print(student_grades)

print("\n" + "="*60)
print("\nSTEP 1: Visualize with boxplots")

fig, ax = plt.subplots(figsize=(10, 5))
student_grades.boxplot(ax=ax)
ax.set_title('Student Performance Distribution by Subject', fontsize=12, fontweight='bold')
ax.set_ylabel('Grade Score')
ax.set_ylim(65, 100)
plt.tight_layout()
plt.show()

print("✓ Boxplots created for comparison")

print("\n" + "="*60)
print("\nSTEP 2: Compare statistics")
for col in student_grades.columns:
    data = student_grades[col]
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    print(f"\n{col}:")
    print(f"  Median: {data.median():.1f}")
    print(f"  Q1-Q3: {Q1:.1f}-{Q3:.1f} (IQR: {IQR:.1f})")
    print(f"  Range: {data.min():.0f}-{data.max():.0f}")

print("\n" + "="*60)
print("\nSTEP 3: Interpret findings")
print("\nMath: Wide box = high variability, grades spread out")
print("  → Some students struggling, others excelling")
print("\nScience: Similar spread to Math, slightly higher median")
print("  → Overall students performing better in Science")
print("\nHistory: Extremely tight box = very consistent")
print("  → All students performing similarly, narrow range")
print("\nConclusion:")
print("→ History grades most predictable")
print("→ Math and Science show more variation")
print("→ Focus support on Math (lower performers)")
```

## 5. End-to-End: Full Boxplot Analysis

```python
temperature_data = pd.DataFrame({
    'City': ['Temp'] * 12,
    'Temperature': [68, 70, 69, 71, 70, 68, 69, 72, -100, 71, 68, 70]
})

print("Temperature Data (with potential outlier):")
print(temperature_data['Temperature'].tolist())

Q1 = temperature_data['Temperature'].quantile(0.25)
Q3 = temperature_data['Temperature'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"\nQuartile Analysis:")
print(f"Q1 (25%): {Q1:.1f}°F")
print(f"Q3 (75%): {Q3:.1f}°F")
print(f"IQR: {IQR:.1f}°F")
print(f"Lower bound: {lower_bound:.1f}°F")
print(f"Upper bound: {upper_bound:.1f}°F")

outliers = temperature_data[(temperature_data['Temperature'] < lower_bound) | 
                            (temperature_data['Temperature'] > upper_bound)]
print(f"\nOutliers detected: {outliers['Temperature'].tolist()}")

print("\n" + "="*60)
print("\nCreating boxplot to visualize outliers...")

fig, ax = plt.subplots(figsize=(6, 4))
ax.boxplot(temperature_data['Temperature'])
ax.set_title('Temperature Distribution with Outlier', fontsize=12, fontweight='bold')
ax.set_ylabel('Temperature (°F)')
plt.tight_layout()
plt.show()

print("✓ Outliers visible as isolated points")
print("\nImportant: -100°F is clearly an error (impossible)")
print("→ Outliers need investigation, not blind removal")
print("→ Ask: Is this a data entry error? Measurement error? Or real?")
print("→ Context matters more than the statistic")
```
