```
def inspect_dataframe(df, name="DataFrame"):
    print(f"\n{'='*60}")
    print(f"INSPECTION REPORT: {name}")
    print(f"{'='*60}\n")
    
    print(f"1. SHAPE: {df.shape}")
    print(f"   Rows: {df.shape[0]} | Columns: {df.shape[1]}\n")
    
    print(f"2. COLUMNS: {df.columns.tolist()}\n")
    
    print(f"3. DATA TYPES:")
    print(df.dtypes)
    print()
    
    print(f"4. MISSING VALUES:")
    print(df.isnull().sum())
    print()
    
    print(f"5. FIRST FEW ROWS:")
    print(df.head())
    print()
    
    print(f"6. BASIC STATISTICS:")
    print(df.describe())
    print(f"{'='*60}\n")
```
