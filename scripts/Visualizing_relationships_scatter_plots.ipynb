## 7. Strategy Guide: When to Use Scatter Plots

```python
print("MISTAKE 1: Confusing scatter plots with line plots")
print("="*60)
print("❌ Wrong: Using scatter plot to show TIME series (day 1, 2, 3...)")
print("✓ Right: Scatter plot = TWO VARIABLES (not time sequence)")
print("Time series → Use LINE plot (shows progression)")
print("Two variables → Use SCATTER plot (shows relationship)")
print("\n" + "="*60 + "\n")

print("MISTAKE 2: Assuming causation from correlation")
print("Observation: More ice cream sales → More drownings")
print("❌ Wrong: 'Ice cream causes drownings!'")
print("✓ Right: 'Both increase in summer (hidden variable: temperature)'")
print("Correlation ≠ Causation")
print("Always ask: Could something else explain both?")
print("\n" + "="*60 + "\n")

print("MISTAKE 3: Using scatter plots on categorical data")
print("❌ Wrong: Scatter plot of [Red, Blue, Green, Yellow]")
print("  (Not numeric - can't plot positions)")
print("✓ Right: Scatter plot of Price ($) vs Quantity")
print("  (Both numeric - can show positions)")
print("\n" + "="*60 + "\n")

print("MISTAKE 4: Ignoring outliers")
print("Data: [10, 12, 11, 9, 13, 10, 1000]")
print("❌ Wrong: Plot it, ignore the 1000")
print("✓ Right: 'Why is 1000 there? Data error? Real event?'")
print("Outliers are signals - investigate, don't dismiss")
print("\n" + "="*60 + "\n")

print("MISTAKE 5: Overinterpreting weak relationships")
print("Few points scattered randomly:")
print("❌ Wrong: 'I see a pattern!'")
print("✓ Right: 'Need more data to confirm pattern'")
print("Weak patterns need strong evidence")
print("Rule: ~20+ points before claiming strong relationship")
```

```python
employee_data = pd.DataFrame({
    'Years Experience': [1, 2, 3, 4, 5, 3.5, 6, 7, 8, 9, 2.5, 5.5, 10, 0.5, 1.5],
    'Annual Salary': [35000, 38000, 42000, 45000, 50000, 44000, 52000, 55000, 60000, 65000, 39000, 51000, 70000, 30000, 36000]
})

print("Employee Experience vs Salary Analysis")
print("="*60)

print("\nSTEP 1: Load data")
print(f"Total employees: {len(employee_data)}")
print(f"Experience range: {employee_data['Years Experience'].min()}-{employee_data['Years Experience'].max()} years")
print(f"Salary range: ${employee_data['Annual Salary'].min():,}-${employee_data['Annual Salary'].max():,}")

print("\n" + "="*60)
print("\nSTEP 2: Create scatter plot")

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(employee_data['Years Experience'], employee_data['Annual Salary'], s=100, alpha=0.6, color='steelblue')
ax.set_title('Employee Experience vs Annual Salary', fontsize=12, fontweight='bold')
ax.set_xlabel('Years of Experience')
ax.set_ylabel('Annual Salary ($)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1000:.0f}K'))
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("✓ Scatter plot created")

print("\n" + "="*60)
print("\nSTEP 3: Analyze patterns")
print("\nRelationship Type:")
if employee_data['Years Experience'].corr(employee_data['Annual Salary']) > 0:
    print("  → POSITIVE: More experience → Higher salary (expected)")
else:
    print("  → NEGATIVE or NO pattern")

print("\nObservations:")
print(f"  • Entry level (0-2 years): ${employee_data[employee_data['Years Experience'] <= 2]['Annual Salary'].mean():,.0f} average")
print(f"  • Mid-level (3-6 years): ${employee_data[(employee_data['Years Experience'] > 2) & (employee_data['Years Experience'] <= 6)]['Annual Salary'].mean():,.0f} average")
print(f"  • Senior (7+ years): ${employee_data[employee_data['Years Experience'] > 6]['Annual Salary'].mean():,.0f} average")

print("\nOutliers:")
potential_outliers = employee_data[
    (employee_data['Years Experience'] > 5) & (employee_data['Annual Salary'] < 45000) |
    (employee_data['Years Experience'] < 3) & (employee_data['Annual Salary'] > 50000)
]
if len(potential_outliers) > 0:
    print("  ⚠ Underpaid senior staff:")
    for idx, row in potential_outliers.iterrows():
        if row['Years Experience'] > 5 and row['Annual Salary'] < 45000:
            print(f"    - {row['Years Experience']} years experience, only ${row['Annual Salary']:,}")
    print("  Action: Review compensation strategy")
else:
    print("  ✓ No major outliers - salaries follow experience")

print("\n" + "="*60)
print("\nSTEP 4: Business Insights")
print("✓ Clear relationship: More experience → Higher pay")
print("✓ No major inequities detected")
print("✓ Progression reasonable: ~$5K per year increase")
print("→ Recommendation: Maintain current pay scale")
```
