```python
import pandas as pd
import numpy as np
```

```python
missing_representations = pd.DataFrame({
    'Original_Value': [1.5, 2.3, 4.1, 5.0, 6.2],
    'None_Type': [1.5, None, 4.1, None, 6.2],
    'NaN_Value': [1.5, np.nan, 4.1, np.nan, 6.2],
    'String_Data': ['Active', 'Active', '', 'Inactive', 'Active']
})

print("Common representations of missing values:")
print(missing_representations)
print("\n" + "="*60 + "\n")

print("Data types:")
print(missing_representations.dtypes)
print("\n" + "="*60 + "\n")

print("How Pandas represents None and np.nan:")
print(f"None value: {None}")
print(f"np.nan value: {np.nan}")
print(f"Are they equal? {np.nan == np.nan}")
print(f"Pandas treats both as NaN (Not a Number)")
```

```python
df = pd.DataFrame({
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Name': ['Alice', np.nan, 'Charlie', 'Diana', np.nan, 'Frank', 'Grace', 'Henry'],
    'Department': ['Sales', 'IT', np.nan, 'IT', 'HR', np.nan, 'Sales', 'IT'],
    'Salary': [50000, 60000, np.nan, 55000, 52000, 65000, np.nan, 58000],
    'Age': [28, 35, 42, np.nan, 38, 45, 31, np.nan],
    'Bonus': [5000, np.nan, 6000, 5500, np.nan, 7000, 6500, np.nan]
})

print("Sample DataFrame with Missing Values:")
print(df)
print("\n" + "="*60 + "\n")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
```

```python
print("Identifying rows with ALL missing values:")
rows_all_missing = df.isnull().all(axis=1)
print(rows_all_missing)
print("\n" + "="*60 + "\n")

print("Checking if missing values exist in the entire DataFrame:")
has_missing = df.isnull().any().any()
print(f"DataFrame has missing values: {has_missing}")
```

```python
print("Percentage of missing values per column:")
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)
print("\n" + "="*60 + "\n")

print("Create a comprehensive missing values report:")
missing_report = pd.DataFrame({
    'Column': df.columns,
    'Missing_Count': df.isnull().sum().values,
    'Missing_Percent': ((df.isnull().sum() / len(df)) * 100).values,
    'Data_Type': df.dtypes.values
})

print(missing_report.to_string(index=False))
```

```python
print("Complete information about row 2:")
print(df.iloc[2])
print("\n" + "="*60 + "\n")

print("Highlight which columns are missing in row 2:")
print(f"Employee_ID missing: {pd.isna(df.iloc[2]['Employee_ID'])}")
print(f"Name missing: {pd.isna(df.iloc[2]['Name'])}")
print(f"Department missing: {pd.isna(df.iloc[2]['Department'])}")
print(f"Salary missing: {pd.isna(df.iloc[2]['Salary'])}")
print(f"Age missing: {pd.isna(df.iloc[2]['Age'])}")
print(f"Bonus missing: {pd.isna(df.iloc[2]['Bonus'])}")
```

```python
print("Columns with high missing percentages (> 20%):")
high_missing = missing_percent[missing_percent > 20]
print(high_missing)
print("\n" + "="*60 + "\n")

print("Columns with no missing values:")
complete_columns = df.columns[df.isnull().sum() == 0]
print(complete_columns.tolist())
```

```python
print("Exercise 4: Find rows with more than 2 missing values")
result_4 = df[df.isnull().sum(axis=1) > 2]
print(result_4)
print("\n" + "="*60 + "\n")

print("Exercise 5: Create a summary showing columns sorted by missing count")
result_5 = df.isnull().sum().sort_values(ascending=False)
print(result_5)
print("\n" + "="*60 + "\n")

print("Exercise 6: Find rows where either Name or Department is missing")
result_6 = df[(df['Name'].isnull()) | (df['Department'].isnull())]
print(result_6)
```

```python
print("Exercise 1: Find total number of missing values in Name column")
result_1 = df['Name'].isnull().sum()
print(f"Total missing in Name: {result_1}")
print("\n" + "="*60 + "\n")

print("Exercise 2: Find which rows have missing Salary")
result_2 = df[df['Salary'].isnull()]
print(result_2)
print("\n" + "="*60 + "\n")

print("Exercise 3: Find percentage of missing Bonus values")
result_3 = (df['Bonus'].isnull().sum() / len(df)) * 100
print(f"Missing Bonus percentage: {result_3:.2f}%")
```

## 7. Key Principles for Missing Value Detection

```python
print("DataFrame with missing information highlighted:")
print(df)
print("\n" + "="*60 + "\n")

print("Visual representation of missing data (True = Missing):")
print(df.isnull())
print("\n" + "="*60 + "\n")

print("Which columns are most affected by missing data:")
missing_severity = df.isnull().sum().sort_values(ascending=False)
print(missing_severity)
```

## 6. Patterns of Missingness

```python
print("Rows where multiple columns are missing (Department or Salary):")
multi_missing = df[df[['Department', 'Salary']].isnull().any(axis=1)]
print(multi_missing)
print("\n" + "="*60 + "\n")

print("Rows where exactly 2 missing values exist:")
exactly_2_missing = df[df.isnull().sum(axis=1) == 2]
print(exactly_2_missing)
```

```python
print("All rows with at least one missing value:")
rows_with_any_missing = df[df.isnull().any(axis=1)]
print(rows_with_any_missing)
print("\n" + "="*60 + "\n")

print("Rows where Name column is missing:")
name_missing_rows = df[df['Name'].isnull()]
print(name_missing_rows)
```

## 5. Inspecting Rows with Missing Data

```python
print("Count of missing values per column:")
missing_per_column = df.isnull().sum()
print(missing_per_column)
print("\n" + "="*60 + "\n")

print("Total missing values in entire DataFrame:")
total_missing = df.isnull().sum().sum()
print(f"Total: {total_missing}")
print("\n" + "="*60 + "\n")

print("Count of missing values per row:")
missing_per_row = df.isnull().sum(axis=1)
print(missing_per_row)
```

## 4. Counting Missing Values

```python
print("Checking missing values in specific columns:")
print("\nName column missing values:")
print(df['Name'].isnull())
print("\n" + "="*60 + "\n")

print("Department column missing values:")
print(df['Department'].isnull())
print("\n" + "="*60 + "\n")

print("Identifying rows with ANY missing values:")
rows_with_missing = df.isnull().any(axis=1)
print(rows_with_missing)
```

```python
print("Boolean mask of missing values using isnull():")
missing_mask = df.isnull()
print(missing_mask)
print("\n" + "="*60 + "\n")

print("Boolean mask using isna() (equivalent to isnull()):")
missing_mask_alt = df.isna()
print(missing_mask_alt)
print("\n" + "="*60 + "\n")

print("Are isnull() and isna() the same?")
print((missing_mask == missing_mask_alt).all().all())
```
