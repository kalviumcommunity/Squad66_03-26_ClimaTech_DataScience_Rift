## 4. Interpreting Outliers Carefully

```python
employee_salary = pd.DataFrame({
    'Salary': [50000, 55000, 52000, 58000, 54000, 53000, 56000, 51000, 
               57000, 52000, 54000, 500000, 53000, 55000, 54000]
})

print("Rule 1: IQR Method (Interquartile Range)")
print("="*60)
print("Data:", employee_salary['Salary'].tolist())

Q1 = employee_salary['Salary'].quantile(0.25)
Q3 = employee_salary['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"\nQ1 (25%): ${Q1:,.0f}")
print(f"Q3 (75%): ${Q3:,.0f}")
print(f"IQR: ${IQR:,.0f}")
print(f"\nOutlier bounds:")
print(f"  Lower: ${lower_bound:,.0f}")
print(f"  Upper: ${upper_bound:,.0f}")

outliers_iqr = employee_salary[
    (employee_salary['Salary'] < lower_bound) | 
    (employee_salary['Salary'] > upper_bound)
]
print(f"\nOutliers detected: {outliers_iqr['Salary'].tolist()}")
print(f"Reason: Exceeds 1.5 * IQR boundary")

print("\n" + "="*60)
print("\nRule 2: Threshold Method")
print("Set reasonable bounds based on domain knowledge")

age_data = pd.DataFrame({
    'Age': [25, 28, 32, 30, 26, 29, 35, 27, 31, 999, 24, 33, 28, 26, 30]
})

print("Data (employees):", age_data['Age'].tolist())
print("\nDomain knowledge:")
print("• Reasonable age range: 18-100")
print("• Any outside range: Data error or mistake")

outliers_threshold = age_data[
    (age_data['Age'] < 18) | (age_data['Age'] > 100)
]
print(f"\nOutliers detected: {outliers_threshold['Age'].tolist()}")
print("Decision: 999 is impossible age → Remove or fix")

print("\n" + "="*60)
print("\nRule 3: Standard Deviation Method")

test_scores = pd.DataFrame({
    'Score': [75, 78, 72, 80, 76, 74, 79, 77, 73, 81, 1000]
})

mean = test_scores['Score'].mean()
std = test_scores['Score'].std()
threshold = 2  # 2 standard deviations

print(f"Data: {test_scores['Score'].tolist()}")
print(f"\nMean: {mean:.1f}")
print(f"Std Dev: {std:.1f}")
print(f"Normal range ({threshold} std): {mean - threshold*std:.1f} to {mean + threshold*std:.1f}")

outliers_std = test_scores[
    (test_scores['Score'] < mean - threshold*std) |
    (test_scores['Score'] > mean + threshold*std)
]
print(f"\nOutliers detected: {outliers_std['Score'].tolist()}")
print("Decision: 1000 score is impossible (max=100) → Data error")

print("\n" + "="*60)
print("\nRule Comparison:")
print("IQR Method: Great for skewed data, robust")
print("Threshold: Simple, requires domain knowledge")
print("Std Dev: Works well for normal distributions")
print("\n→ Best practice: Use multiple methods together")
```
