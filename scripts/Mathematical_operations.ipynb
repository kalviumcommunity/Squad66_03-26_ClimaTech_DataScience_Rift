```python
import numpy as np
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([1, 2, 3, 4])

print("Array 1:", arr1)
print("Array 2:", arr2)

print("\n-----------------------\n")

print("Addition:", arr1 + arr2)
print("Subtraction:", arr1 - arr2)
print("Multiplication:", arr1 * arr2)
print("Division:", arr1 / arr2)

print("\n-----------------------\n")

print("Add scalar ( + 5 ):", arr1 + 5)
print("Multiply scalar ( * 2 ):", arr1 * 2)
print("Divide scalar ( / 2 ):", arr1 / 2)

print("\n-----------------------\n")

list1 = [10, 20, 30, 40]

print("Python list:", list1)

print("List + list (concatenation):", list1 + list1)
print("NumPy array + array:", arr1 + arr2)

print("\n-----------------------\n")

print("NumPy makes math element-wise and simple!")
```
