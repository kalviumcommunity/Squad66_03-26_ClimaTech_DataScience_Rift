## 6. End-to-End: Full Distribution Comparison

```python
income_data = pd.DataFrame({
    'Tech Salaries': [60000, 70000, 75000, 80000, 85000, 90000, 500000],
    'Retail Salaries': [25000, 26000, 27000, 28000, 29000, 30000, 31000],
    'Mixed Salaries': [30000, 35000, 40000, 45000, 50000, 100000, 200000]
})

print("Salary Data (different industries):")
print(income_data)
print("\n" + "="*60)

print("\nStep 1: Detect skewness by comparing mean vs median")
for col in income_data.columns:
    mean_val = income_data[col].mean()
    median_val = income_data[col].median()
    diff = mean_val - median_val
    
    print(f"\n{col}:")
    print(f"  Mean: ${mean_val:,.0f}")
    print(f"  Median: ${median_val:,.0f}")
    print(f"  Difference: ${diff:,.0f}")
    
    if abs(diff) < 5000:
        print(f"  → Symmetric distribution (mean ≈ median)")
    elif diff > 0:
        print(f"  → Right-skewed (high outliers pull mean up)")
    else:
        print(f"  → Left-skewed (low outliers pull mean down)")

print("\n" + "="*60)
print("\nStep 2: Understand what skewness means")
print("\nTech Salaries:")
print("  Most earn $60-90K, but one executive earns $500K")
print("  Mean is pulled HIGH by the outlier")
print("  → Not representative of typical employee!")

print("\nRetail Salaries:")
print("  Most earn $25-31K, no huge outliers")
print("  Mean close to median")
print("  → Mean IS representative!")

print("\nMixed Salaries:")
print("  Some earn $30-50K, others $100-200K")
print("  Multiple high outliers")
print("  → Highly skewed, bimodal distribution!")
```

## 4. Comparing Quartiles and Percentiles

```python
temperature_data = pd.DataFrame({
    'City A': [70, 71, 69, 72, 70, 71],
    'City B': [50, 80, 60, 90, 55, 85],
    'City C': [65, 66, 64, 67, 65, 66]
})

print("Temperature Data (multiple locations):")
print(temperature_data)
print("\n" + "="*60)

print("\nStep 1: Compare ranges")
for col in temperature_data.columns:
    data_range = temperature_data[col].max() - temperature_data[col].min()
    print(f"{col}: {temperature_data[col].min()}°F to {temperature_data[col].max()}°F (range: {data_range}°F)")

print("\n" + "="*60)
print("\nStep 2: Compare standard deviation (variability)")
std_values = temperature_data.std()
print(f"\nVariability (Std Dev):")
for col in temperature_data.columns:
    print(f"  {col}: {std_values[col]:.2f}°F")

print(f"\nMost stable: {std_values.idxmin()} ({std_values.min():.2f}°F std dev)")
print(f"Most variable: {std_values.idxmax()} ({std_values.max():.2f}°F std dev)")

print("\n" + "="*60)
print("\nStep 3: Compare coefficient of variation (variability relative to mean)")
print("Formula: std dev / mean × 100 (lower % = more stable relative to average)")
for col in temperature_data.columns:
    cv = (temperature_data[col].std() / temperature_data[col].mean()) * 100
    print(f"{col}: {cv:.1f}% coefficient of variation")

print("\nInterpretation:")
print("City A: Stable temperature, predictable")
print("City B: Highly variable, unpredictable")
print("City C: Very stable, most predictable")
```

```python
store_data = pd.DataFrame({
    'Store A': [100, 105, 95, 110, 100, 98],
    'Store B': [200, 180, 220, 190, 210, 200],
    'Store C': [80, 85, 75, 90, 82, 80]
})

print("Daily Sales by Store:")
print(store_data)
print("\n" + "="*60)

print("\nStep 1: Get summary statistics for all columns")
stats = store_data.describe().T[['mean', 'median', 'min', 'max']]
print(stats)

print("\n" + "="*60)
print("\nStep 2: Compare mean values")
means = store_data.mean()
print(f"\nAverage Daily Sales:")
for col in store_data.columns:
    print(f"  {col}: ${means[col]:,.0f}")

print(f"\nHighest average: {means.idxmax()} at ${means.max():,.0f}")
print(f"Lowest average: {means.idxmin()} at ${means.min():,.0f}")
print(f"Difference: ${means.max() - means.min():,.0f}")

print("\n" + "="*60)
print("\nStep 3: Compare mean vs median (detect skew)")
for col in store_data.columns:
    mean_val = store_data[col].mean()
    median_val = store_data[col].median()
    skew = mean_val - median_val
    print(f"{col}: Mean ${mean_val:.0f} vs Median ${median_val:.0f} (diff: ${skew:.0f})")
    if abs(skew) < 5:
        print(f"  → Fairly symmetric")
    else:
        print(f"  → {('High' if skew > 0 else 'Low')} outliers affecting mean")
```
